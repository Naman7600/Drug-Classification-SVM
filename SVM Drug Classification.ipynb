```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for i in df.columns:
    if df[i].dtype=='object':
        df[i] = le.fit_transform(df[i])
```

```python
df
```

```python
x=df.drop("Drug",axis=1)
y=df["Drug"]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=62)
```

```python
from sklearn.svm import SVC
```

```python
model1=SVC()
```

```python
model1.fit(x_train,y_train)

y1_pred=model1.predict(x_test)
```

```python
y1_pred
```

```python
from sklearn.metrics import * 
```

```python
print(f"Accuracy Score is : {accuracy_score(y1_pred,y_test)}")
```

```python
model2=SVC(kernel='linear')
model2.fit(x_train,y_train)

y2_pred=model2.predict(x_test)
```

```python
print(f"Accuracy Score is {accuracy_score(y2_pred,y_test)}")
```

```python
model3=SVC(kernel='poly')
model3.fit(x_train,y_train)

y3_pred=model3.predict(x_test)
```

```python
print(f"Accuracy Score is {accuracy_score(y3_pred,y_test)}")
```

```python
model4=SVC(kernel='sigmoid')
model4.fit(x_train,y_train)

y4_pred=model4.predict(x_test)
```

```python
print(f"Accuracy Score is {accuracy_score(y4_pred,y_test)}")
```
